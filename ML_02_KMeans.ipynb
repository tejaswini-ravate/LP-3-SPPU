```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.shape
```

```python
features = ['SALES','QUANTITYORDERED','PRICEEACH']
```

```python
X = df[features]
```

```python
sc = StandardScaler()
```

```python
X_scaled = pd.DataFrame(sc.fit_transform(X),columns = features)
```

```python
inertia = []
k_range = range(1,11)

for k in k_range:
    km = KMeans(n_clusters = k, random_state = 42)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.plot(k_range,inertia,marker = 'o')
```

```python
k = 3
km = KMeans(n_clusters = k, random_state = 42)
clusters = km.fit_predict(X_scaled)

df['kmeans_cluster'] = clusters
centers = pd.DataFrame(km.cluster_centers_,columns=features)
centers
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage,fcluster
```

```python
Z = linkage(X_scaled,method = 'ward')
```

```python
plt.figure(figsize = (10,6))
dendrogram(Z,truncate_mode = 'lastp',p = 30, leaf_rotation = 90.,leaf_font_size = 12.,show_contracted = True);
```

```python
max_d = 7
cluster_h = fcluster(Z,t = 4,criterion = 'maxclust')
df['hier_cluster'] = cluster_h
```

```python
print("K means count:")
print(df['kmeans_cluster'].value_counts())

print("\n Hierarchical Clustering")
print(df['hier_cluster'].value_counts())
```

```python
pd.crosstab(df['kmeans_cluster'],df['hier_cluster'])
```

```python
from sklearn.metrics import adjusted_rand_score
```

```python
ari = adjusted_rand_score(df['kmeans_cluster'],df['hier_cluster'])
```
