```python
def frac_knapsack(val,wt,W):
    n = len(val)
    items = []

    for i in range(n):
        items.append((val[i]/wt[i],wt[i],val[i]))

    items.sort()

    rem_cap = W
    total_val = 0

    for ratio,weight,val in items:
        if weight <= rem_cap:
            total_val += val
            rem_cap -= weight
        else:
            total_val += (rem_cap/weight)*val
            rem_cap = 0
            break
    return total_val
```

```python
def main():
    while True:
        val = list(map(int,input("Enter values:").split()))
        wt = list(map(int,input("Enter weights:").split()))
        W = int(input("Enter capacity:"))
        print(f"Maximum Profit:{frac_knapsack(val,wt,W)}")
    
        ans = str(input("Enter y/n to continue?:"))
        if ans == 'n':
            break
    
        print()
```

```python
if __name__ == "__main__":
    main()
```
